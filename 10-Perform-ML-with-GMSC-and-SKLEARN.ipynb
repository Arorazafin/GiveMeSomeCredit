**Description of the output and features**  
Features indicated on the website are all prensent the csv file.  
We show a quick descriptive on features :
- **Output : 'SeriousDlqin2yrs'** : Person experienced 90 days past due delinquency or worse 
- **Feature 1 : 'RevolvingUtilizationOfUnsecuredLines'**	Total balance on credit cards and personal lines of credit except real estate and no installment debt like car loans divided by the sum of credit limits
- **Feature 2 : 'age'**	Age of borrower in years
- **Feature 3  : 'NumberOfTime30-59DaysPastDueNotWorse'**	Number of times borrower has been 30-59 days past due but no worse in the last 2 years.
- **Feature 4 : 'DebtRatio'**	Monthly debt payments, alimony,living costs divided by monthy gross income
- **Feature 5 : 'MonthlyIncome'**	Monthly income
- **Feature 6 : 'NumberOfOpenCreditLinesAndLoans'**	Number of Open loans (installment like car loan or mortgage) and Lines of credit (e.g. credit cards)
- **Feature 7 : 'NumberOfTimes90DaysLate'**	Number of times borrower has been 90 days or more past due.
- **Feature 8 : 'NumberRealEstateLoansOrLines'**	Number of mortgage and real estate loans including home equity lines of credit
- **Feature 10 : 'NumberOfTime60-89DaysPastDueNotWorse'**	Number of times borrower has been 60-89 days past due but no worse in the last 2 years.
- **Feature 11 : 'NumberOfDependents'**	Number of dependents in family excluding themselves (spouse, children etc.)

### Representation of Monthtly Income and Number of dependents by Age 

```python
cols = list(df_train_full.columns.values)
col_to_consider = ['MonthlyIncome','NumberOfDependents']

grouped_df = df_train_full.groupby('age')
    
i = 1
plt.figure(figsize = (40,40))

for col in col_to_consider:
    plt.subplot(5,5,i)
    data_by_age = grouped_df[col].aggregate([np.mean,'count']).reset_index()
    data_by_age.columns =['age',col,'count']
    sns.regplot(x='age',y=col,data = data_by_age , order = 2)
    plt.xlabel(col)    
    i=i+1
    #plt.xticks(rotation='45',size=8)
    
plt.show()
data_by_age.head(5)
```

**Analysis:**
Regarding Monthly income
There is an existense of outliers :
- Age 0 has a MonthyIncome of 6.000
- Age 105, 107, 109 have a missing value Monthly income 
=> Create a new feature : MonthlyIncome_by_average
This supposed represented our sample


### New feature 1 : MonthlyIncome_average_by_age

```python
MonthlyIncome_by_age = grouped_df["MonthlyIncome"].aggregate([np.mean]).reset_index()
#rename the column to "mean"
col_names = MonthlyIncome_by_age.columns.values
col_names[1] = 'MonthlyIncome_average_by_age'
MonthlyIncome_by_age.columns = col_names
print(MonthlyIncome_by_age.head (5))
print()

#Creation of v1
df_train_full_v1 = df_train_full.copy()

df_train_full_v1 = df_train_full_v1.join(MonthlyIncome_by_age.set_index('age'), on='age')
df_train_full_v1.info()
df_train_full_v1[df_train_full_v1["MonthlyIncome_average_by_age"].isnull()]

```

df_train_full_v1 will be the dataframe with a new features :
- age_mod, where the 0 age will change
- Monthly_Income_average_ by_age, and the MonthlyIncome average for age > 105 will be 0

```python
def_inter = df_train_full_v1[(df_train_full_v1['age'] != 0) & (df_train_full_v1['MonthlyIncome_average_by_age'] > 5990) &  (df_train_full_v1['MonthlyIncome_average_by_age'] < 6035)]
age_mi6000 = def_inter.age.iloc [0]
print('age for MonthlyIcome average of 6000 : ',age_mi6000)
df_train_full_v1['age_mod'] = df_train_full_v1['age']
df_train_full_v1.loc[df_train_full_v1['age_mod']==0,'age_mod'] = age_mi6000
df_train_full_v1.loc[df_train_full_v1['age'] >= 105,'MonthlyIncome_average_by_age'] = 0


df_train_full_v1.info()

df_train_full_v1_qgrid = qgrid.show_grid(df_train_full_v1)
df_train_full_v1_qgrid
```

Dataframe v1 shows new features 'age_mod' and 'Monthly_Income_average_ by_age' created with no NaN value and outlier age zero has been smoothed on new features

### Select features to transforn into log

```python
# transformation into log
df_train_full_log = df_train_full.copy() 

#df_train_full.head(5)
#df_train_full_log.head(6)


trans_log = ['RevolvingUtilizationOfUnsecuredLines',
            'NumberOfTime30-59DaysPastDueNotWorse',
            'DebtRatio',
            'NumberOfTimes90DaysLate',
            'NumberOfOpenCreditLinesAndLoans', 
            'NumberRealEstateLoansOrLines',
            'NumberOfTime60-89DaysPastDueNotWorse']

for c in trans_log:
    df_train_full_log[c] = np.log(1 + df_train_full[c])                                    
    
    
cols = list(df_train_full_log.columns.values)
remove_col = ['Id','SeriousDlqin2yrs','MonthlyIncome','NumberOfDependents']
for rm in remove_col:
    cols.remove(rm)
    
i = 1
plt.figure(figsize = (20,10))

for col in cols:
    plt.subplot(2,4,i)
    sns.distplot(df_train_full_log[col])
    plt.xlabel(col)    
    i=i+1
    plt.xticks(rotation='45',size=8)
plt.show()
    
    
```

### New features 3 : features transformed to log  

```python
df_train_full_v3 = df_train_full_v2.copy()

trans_log = ['RevolvingUtilizationOfUnsecuredLines',
            'NumberOfTime30-59DaysPastDueNotWorse',
            'DebtRatio',
            'NumberOfTimes90DaysLate',
            'NumberRealEstateLoansOrLines',
            'NumberOfOpenCreditLinesAndLoans', 
            'NumberOfTime60-89DaysPastDueNotWorse']

for c in trans_log:
    new_feat = c+"_log"
    df_train_full_v3[new_feat] = np.log(1 + df_train_full_v3[c])                                    
    

df_train_full_v3.info()


#Visualizing DataFrame
df_train_full_v3_qgrid = qgrid.show_grid(df_train_full_v3)
df_train_full_v3_qgrid
```

## New features 4 : intituive new features

let's try to add this new features, which are 
- IncomePerPerson
- NumOfPastDue
- MonthlyDebt
- NumOfOpenCreditLines
- MonthlyBalance
then we noramlize in with logarythm method

```python
# NaN can not works with calulation. So we replace that in zero.
#df_train_full_v0 = df_train_full.fillna(0)

# New features no normalized
df_train_full_v4 = df_train_full_v3.fillna(0).copy()

df_train_full_v4['IncomePerPerson'] = df_train_full_v4['MonthlyIncome']/(df_train_full_v4['NumberOfDependents']+1)

df_train_full_v4['NumberOfTimesDaysLateCumul'] = df_train_full_v4['NumberOfTimes90DaysLate']+df_train_full_v4['NumberOfTime60-89DaysPastDueNotWorse'] +df_train_full_v4['NumberOfTime30-59DaysPastDueNotWorse']

df_train_full_v4['MonthlyDebt'] = df_train_full_v4['DebtRatio']*df_train_full_v4['MonthlyIncome']

df_train_full_v4['NumOfOpenCreditLinesCumul'] = df_train_full_v4['NumberOfOpenCreditLinesAndLoans'] + df_train_full_v4['NumberRealEstateLoansOrLines']

df_train_full_v4['MonthlyBalance'] = df_train_full_v4['MonthlyIncome'] - df_train_full_v4['MonthlyDebt']

df_train_full_v4.info()

```

```python
# New features "normalized" transorformed into log
df_train_full_v5 = df_train_full_v4.copy()

df_train_full_v5['IncomePerPerson_log'] = np.log(1+df_train_full_v5['IncomePerPerson'])
df_train_full_v5['NumberOfTimesDaysLateCumul_log'] = np.log(1+df_train_full_v5['NumberOfTimesDaysLateCumul'])

df_train_full_v5['MonthlyDebt_log'] = np.log(1+df_train_full_v5['MonthlyDebt'])

df_train_full_v5['NumOfOpenCreditLinesCumul_log'] = np.log(1+df_train_full_v5['NumOfOpenCreditLinesCumul'])

val_min_abs = abs(df_train_full_v5['MonthlyBalance'].min())   
df_train_full_v5['MonthlyBalance_log'] = np.log(1+(df_train_full_v5['MonthlyBalance'] + val_min_abs))

df_train_full_v5.info()
```

```python
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split

from sklearn.metrics import roc_curve, roc_auc_score #AUC score
```

```python
def forest_roc_curve(fpr, tpr, label=None):
    plt.figure(figsize=(12,10))
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0,1],[0,1], "k--")
    plt.axis([0,1,0,1])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive rate")
```

```python
def model_feature_importance(x_data,model):
    feature_imp = pd.Series(model.feature_importances_,index=x_data.columns).sort_values(ascending=False)
    # Creating a bar plot
    sns.barplot(x=feature_imp, y=feature_imp.index)
    # Add labels to your graph
    plt.xlabel('Feature Importance Score')
    plt.ylabel('Features')
    plt.title("Visualizing Important Features")
    plt.show()
```

```python
model = "random_forest"
max_depth_max = 13
#parameter for the model (lower 6)


dastaset_simul = pd.DataFrame({'name': ['data_orig', 'data_orig_aver', 'data_orig_newf','data_log_aver','data_log_aver_lognewf', 'data_log_aver_lognewf'],
                               'dataset' :[df_train_full_vf0,df_train_full_vf1,df_train_full_vf2,df_train_full_vf3,df_train_full_vf4,df_train_full_vf5]})

auc_sim = pd.DataFrame({'model': [0], 'Dataset': [0],'max_depth':[0],'AUC X_train':[0],'AUC X_val':[0],'difference':[0]})
                                       
                            


for sim in range(0,5):

    # data
    df_train_for_model = dastaset_simul.iloc[sim,1].copy()
    df_train_for_model = df_train_for_model.drop('Id', axis=1)
    #df_train_for_model.info()
    #print()

    #Split
    val_size = 0.2
    x_train, x_val = train_test_split(df_train_for_model, test_size = val_size)
    #print("X Train full shape: " + str(df_train_for_model.shape))

    #Separated x and y
    target_column = 'SeriousDlqin2yrs'
    y_train = x_train[target_column]
    x_train = x_train.drop(target_column, axis=1)
    #print("X Train shape: " + str(x_train.shape))

    y_val = x_val[target_column]
    x_val = x_val.drop(target_column, axis=1)
    #print("X Val shape: " + str(x_val.shape))
    #print()
    
    
    
    for d in range(5,max_depth_max):
        # Create model
        forest = RandomForestClassifier(random_state=42,n_estimators=300, max_depth=d, class_weight="balanced")
        forest.fit(x_train,y_train) 
    

        #Evaluation
        y_train_proba = forest.predict_proba(x_train)
        y_scores = y_train_proba[:,1]
        roc_xTrain = roc_auc_score(y_train,y_scores)
        #cross validation
        y_val_proba = forest.predict_proba(x_val)
        y_val_scores = y_val_proba[:,1]
        roc_xVal = roc_auc_score(y_val,y_val_scores)
        auc_sim = auc_sim.append({'model': model, 'Dataset': dastaset_simul.iloc[sim,0] ,'max_depth':d,'AUC X_train':roc_xTrain,'AUC X_val':roc_xVal, 'difference' :roc_xVal-roc_xTrain }, ignore_index = True)

print(auc_sim)
print()
df_auc_max = auc_sim[auc_sim['AUC X_val'] == auc_sim['AUC X_val'].max()]
print("Optimal AUC : ") 
print(df_auc_max)
 

#save file
namefile = model+"_auc_sim.csv"
auc_sim.to_csv(namefile)
```

## Learning rate parameter

```python
model = "XGBClassifier"
#max_depth_max = 5
#let's change learning_rate

learnin_rate_chg = [0.01,0.015,0.02,0.025,0.03]

dastaset_simul = df_train_full_vf1.copy()

auc_sim = pd.DataFrame({'learning_rate':[0],'AUC X_train':[0],'AUC X_val':[0],'difference':[0]})
                                       
                            



# data
df_train_for_model = dastaset_simul.copy()
df_train_for_model = df_train_for_model.drop('Id', axis=1)
#df_train_for_model.info()
#print()

#Split
val_size = 0.2
x_train, x_val = train_test_split(df_train_for_model, test_size = val_size)
#print("X Train full shape: " + str(df_train_for_model.shape))

#Separated x and y
target_column = 'SeriousDlqin2yrs'
y_train = x_train[target_column]
x_train = x_train.drop(target_column, axis=1)
#print("X Train shape: " + str(x_train.shape))

y_val = x_val[target_column]
x_val = x_val.drop(target_column, axis=1)
#print("X Val shape: " + str(x_val.shape))
#print()
       
for lr in learnin_rate_chg:
    # Create model
    xgb_model = XGBClassifier(n_estimators=550, max_depth=5, learning_rate=lr,  subsample=0.5, colsample_bytree=0.5)
    xgb_model.fit(x_train,y_train) 
    
    #Evaluation
    y_train_proba = xgb_model.predict_proba(x_train)
    y_scores = y_train_proba[:,1]
    roc_xTrain = roc_auc_score(y_train,y_scores)
    #cross validation
    y_val_proba = xgb_model.predict_proba(x_val)
    y_val_scores = y_val_proba[:,1]
        
    #score
    roc_xVal = roc_auc_score(y_val,y_val_scores)
    auc_sim = auc_sim.append({'learning_rate':lr,'AUC X_train':roc_xTrain,'AUC X_val':roc_xVal, 'difference' :roc_xVal-roc_xTrain }, ignore_index = True)

print(auc_sim)
print()
df_auc_max = auc_sim[auc_sim['AUC X_val'] == auc_sim['AUC X_val'].max()]
print("Optimal AUC : ") 
print(df_auc_max)
 
```
